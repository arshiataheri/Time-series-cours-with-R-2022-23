```r
xdiff = diff(x)
plot(xdiff)
# The differenciated one seems stationary cause we can remark a constant average and variance.

library(urca)
summary(ur.df(xdiff))

# Dickey Fullet test : hypothèse nulle = présene d'une racine unitaire = processus non stationnaire
# la valeur absolue de la statistique (23.82 ) est bien superieur à la valeur absolues de référence à 1%, (2.58) on rejete l'hypothèse nulle, la série est stationnaire
```

```r
library(forecast)
bestarima = arima(xdiff, order=c(1,0,2))
checkresiduals(bestarima)
# The first plot show the res vs. time and it's clearly obvious that there exists no clear trends and tendencies in the residuals so they can be considered as white noise.

# Below at left the plot exhibits an auto-corr and a partial auto-corr at Lag number 4.

# Below at rights show if the residuals are normaly distributed or not. Which is not(clearly), here.


res = residuals(bestarima)
Box.test(res, lag=1, type = "Ljung")
# We choose a lag of 1, because we want to see if there is autocorrelation with each lag.
Box.test(res, lag=12, type = "Ljung")
# Now, we conduct another case with lag 12, to see if the time series have seasonality every year.


# analyse de lung-box : h0 pas d'autocorrélation . si pvalue inf à 5% on accepte h0. Dans notre cas la p-value est inf à 5% donc on accepte H0 il y a pas d'autocorrélation

# The p-value for the test with lag 1 is 0.80, which is greater than the significance level of 0.05. Therefore, we do not reject the null hypothesis, and we conclude that there is no evidence of autocorrelation in the dataset up to lag 1.

# FOr lag = 12, P-value is 0.15 and is more than 0.05. Therefore, we accept the null hypothesis and conclude that there is no evidence of autocorrelation in the dataset up to lag 12.


# NOTE
# However, the second test with a lag of 12 indicates that there is evidence of autocorrelation at lag 12, which suggests the presence of seasonality in the time series. The presence of seasonality means that the time series exhibits a regular pattern that repeats at fixed intervals (e.g., every 12 months).
```
